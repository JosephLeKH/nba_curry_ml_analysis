```python
import pandas as pd

df = pd.read_csv('FINAL_Steph.csv')
standings_df = pd.read_csv('FINAL_nba_standings.csv')

df.drop('Unnamed: 0', axis=1, inplace=True)
df
```

```python
df.columns
```

```python
df['Home'] = df['MATCHUP'].str.contains('vs.')
df['Home']
```

```python
df['MIN'] = df['MIN'].replace(0, 0.00)
df['MIN'] = df['MIN'].str.split(":").str[0].astype(float)
df['MIN']
```

```python
df['Opp'] = df['MATCHUP'].str.split(' ').str[2]
df['Opp']
```

```python
df['Steph'] = df['MIN'] > 0
df['Steph']
```

```python
df['Win'] = df['WL'].str.contains('W')
df['Win']
```

```python
# Clean team name function
def clean_team_name(name):
    import re
    name = re.sub(r'\*|\xa0|\s*\(\d+\)', '', name).strip()
    return name

# Apply cleaning
standings_df['Team'] = standings_df['Team'].apply(clean_team_name)

# Full team abbreviation mapping (feel free to expand or edit as needed)
team_abbreviations = {
    'Atlanta Hawks': 'ATL',
    'Boston Celtics': 'BOS',
    'Brooklyn Nets': 'BKN',
    'Charlotte Hornets': 'CHA',
    'Charlotte Bobcats': 'CHA',  # historical name
    'Chicago Bulls': 'CHI',
    'Cleveland Cavaliers': 'CLE',
    'Dallas Mavericks': 'DAL',
    'Denver Nuggets': 'DEN',
    'Detroit Pistons': 'DET',
    'Golden State Warriors': 'GSW',
    'Houston Rockets': 'HOU',
    'Indiana Pacers': 'IND',
    'Los Angeles Clippers': 'LAC',
    'Los Angeles Lakers': 'LAL',
    'Memphis Grizzlies': 'MEM',
    'Miami Heat': 'MIA',
    'Milwaukee Bucks': 'MIL',
    'Minnesota Timberwolves': 'MIN',
    'New Jersey Nets': 'NJN',  # historical name
    'New Orleans Hornets': 'NOH',  # historical name
    'New Orleans Pelicans': 'NOP',
    'New York Knicks': 'NYK',
    'Oklahoma City Thunder': 'OKC',
    'Orlando Magic': 'ORL',
    'Philadelphia 76ers': 'PHI',
    'Phoenix Suns': 'PHX',
    'Portland Trail Blazers': 'POR',
    'Sacramento Kings': 'SAC',
    'San Antonio Spurs': 'SAS',
    'Toronto Raptors': 'TOR',
    'Utah Jazz': 'UTA',
    'Washington Wizards': 'WAS'
}

# Map to abbreviation
standings_df['Team'] = standings_df['Team'].map(team_abbreviations)

standings_df
```

```python
standings_df[standings_df['Season'] == '2009-10']['W/L%'].sort_values(ascending=False)
```

```python
# First, rename the columns in second_df to prepare for merge
team_strength_df = standings_df.rename(columns={
    'Team': 'Opp',
    'Season': 'SEASON'
})

# Merge on Opp and SEASON
df = df.merge(team_strength_df, on=['Opp', 'SEASON'], how='left')

# Create the boolean flag for strong opponents
df['Strong'] = df['W/L%'] > 0.6

```

```python
df.to_csv('CS109_Final_Project')
```
